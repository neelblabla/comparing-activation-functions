## AFs Visualization

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set the Seaborn white grid style
sns.set_style('whitegrid')

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate some input values
x = np.linspace(-10, 10, 100)

# Compute the output values
y = sigmoid(x)

plt.figure(figsize=(10, 5))

# Plot the function
plt.plot(x, y)

# Add title and clear labels
plt.title('')
plt.xticks(size=18)
plt.yticks(size=18)
plt.xlabel('')

# Show the plot
plt.show()
```

```python
# Define the ReLU function
def relu(x):
    return np.maximum(0, x)

# Generate some input values
x = np.linspace(-10, 10, 100)

# Compute the output values
y = relu(x)


plt.figure(figsize=(10, 5))

# Plot the function
plt.plot(x, y)

# Add title and clear labels
plt.title('')
plt.xticks(size=18)
plt.yticks(size=18)
plt.xlabel('')

# Show the plot
plt.show()
```

```python
# Define the Leaky ReLU function
def leaky_relu(x, alpha=0.1):
    return np.maximum(alpha * x, x)

# Generate some input values
x = np.linspace(-10, 10, 100)

# Compute the output values
y = leaky_relu(x)


plt.figure(figsize=(10, 5))

# Plot the function
plt.plot(x, y)

# Add title and clear labels
plt.title('')
plt.xticks(size=18)
plt.yticks(size=18)
plt.xlabel('')

# Show the plot
plt.show()
```

```python
# Define the hyperbolic tangent (tanh) function
def tanh(x):
    return np.tanh(x)

# Generate some input values
x = np.linspace(-10, 10, 100)

# Compute the output values
y = tanh(x)


plt.figure(figsize=(10, 5))

# Plot the function
plt.plot(x, y)

# Add title and clear labels
plt.title('')
plt.xticks(size=18)
plt.yticks(size=18)
plt.xlabel('')

# Show the plot
plt.show()
```

```python
# Softmax computes the probability of each output node so it is not really possible to show a general
# plot of it, sohere is just the exponential x divided by sum of exponential x

# Define the softmax function
def softmax(x):
    exp_x = np.exp(x)
    return exp_x / np.sum(exp_x)

# Generate some input values
x = np.linspace(-10, 10, 100)

# Compute the output values
y = softmax(x)

plt.figure(figsize=(10, 5))

# Plot the function
plt.plot(x, y)

# Add title and clear labels
plt.title('')
plt.xticks(size=18)
plt.yticks(size=18)
plt.xlabel('')
plt.ylabel('')

# Show the plot
plt.show()
```

```python
# Define the exponential linear unit (ELU) function
def elu(x, alpha=1.0):
    return np.where(x >= 0, x, alpha * (np.exp(x) - 1))

# Generate some input values
x = np.linspace(-10, 10, 100)

# Compute the output values
y = elu(x)

plt.figure(figsize=(10, 5))

# Plot the function
plt.plot(x, y)

# Add title and clear labels
plt.title('')
plt.xticks(size=18)
plt.yticks(size=18)
plt.xlabel('')

# Show the plot
plt.show()
```

```python
import numpy as np

# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# ReLU function
def relu(x):
    return np.maximum(0, x)

# Leaky ReLU function
def leaky_relu(x, alpha=0.01):
    return np.maximum(alpha * x, x)

# Hyperbolic tangent function
def tanh(x):
    return np.tanh(x)

# Softmax function
def softmax(x):
    exps = np.exp(x - np.max(x))
    return exps / np.sum(exps)

# ELU function
def elu(x, alpha=1.0):
    return np.where(x >= 0, x, alpha * (np.exp(x) - 1))

# Print equations
print("Equations for activation functions:\n")
print("Sigmoid: f(x) = 1 / (1 + exp(-x))")
print("ReLU: f(x) = max(0, x)")
print("Leaky ReLU: f(x) = max(alpha * x, x) where alpha is a small positive constant")
print("Hyperbolic Tangent (tanh): f(x) = tanh(x)")
print("Softmax: f(x) = exp(x) / sum(exp(x))")
print("Exponential Linear Unit (ELU): f(x) = x if x >= 0 else alpha * (exp(x) - 1) where alpha is a small positive constant")
```

## State of the art AFs comparison

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')

# Define the activation functions
def swish(x, beta=1):
    return x / (1 + np.exp(-beta * x))

def mish(x):
    return x * np.tanh(np.log(1 + np.exp(x)))

def selu(x, alpha=1.67326, scale=1.0507):
    return scale * np.where(x < 0, alpha * (np.exp(x) - 1), x)

def gelu(x, alpha=1.702):
    return x * 0.5 * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + alpha * (x**3))))

def aelu(x, alpha=1.5):
    return np.where(x < 0, alpha * (np.exp(x) - 1), x * alpha + x)

# Set up the input range
x = np.linspace(-5, 5, 1000)

# Compute the activation functions
y_swish = swish(x)
y_mish = mish(x)
y_selu = selu(x)
y_gelu = gelu(x)
y_aelu = aelu(x)

# Plot the activation functions
#sns.set_style("darkgrid")
#sns.set_palette("bright")
plt.figure(figsize=(20, 13))
plt.plot(x, y_swish, label="Swish", color = 'blue', linewidth=3, linestyle='-')
plt.plot(x, y_mish, label="Mish", color = 'red', linewidth=3, linestyle='-')
plt.plot(x, y_selu, label="SELU", color = 'black', linewidth=3, linestyle='-')
plt.plot(x, y_gelu, label="GELU", color = 'green', linewidth=3, linestyle='-')
plt.plot(x, y_aelu, label="AELU", color='brown', linestyle='--', linewidth=3)
plt.legend(fontsize=30)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.show()
```
