```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def binaryStep(x):
    ''' It returns '0' is the input is less then zero otherwise it returns one '''
    return np.heaviside(x,1)
```

```python
x = np.linspace(-10, 10)
plt.plot(x, binaryStep(x))
plt.axis('tight')
plt.title('Binary Step')
plt.grid(alpha=0.4, ls='-.')
plt.show()
```

```python
def sigmoid(x):
    ''' It returns 1/(1+exp(-x)). where the values lies between zero and one '''

    return 1/(1+np.exp(-x))
```

```python
x = np.linspace(-10, 10)
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.title('Sigmoid')
plt.grid(alpha=0.4, ls='-.')
plt.show()
```

```python
def RELU(x):
    ''' It returns zero if the input is less than zero otherwise it returns the given input. '''
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)

    return x1
```

```python
x = np.linspace(-10, 10)
plt.plot(x, RELU(x))
plt.axis('tight')
plt.title('ReLU')
plt.grid(alpha=0.4, ls='-.')
plt.show()
```

```python
def plot(func, yaxis=(-0.5, 2)):
    plt.ylim(yaxis)
    plt.title('ELU')
    plt.grid(alpha=0.4, ls='-.')
    plt.plot(x, func(x))
```

```python
elu = np.vectorize(lambda x: x if x > 0 else 0.5 * (np.exp(x) - 1), otypes=[np.float])
```
